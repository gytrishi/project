0+1+2 survey data code 4w

```python
url1="/content/surveyZero_one_twoFaceToFace_csvutf_4w.csv"
url2="/content/cardlist4wlc.csv"
url3="/content/mnldata4w.csv"
url4="/content/cardlist4wvc.csv"
import pandas as pd
import numpy as np
import math
d1=pd.read_csv(url1)
d2=pd.read_csv(url2)
d3=pd.read_csv(url3)
d4=pd.read_csv(url4)
#df_VC_modification
df_VC=d1.iloc[:,7:78]
header=[i for i in range(0,df_VC.shape[1])]
header
df_VC.columns=header
df_VC.columns
#df_LC_modification
df_LC=d1.iloc[:,93:271]
header=[i for i in range(0,df_LC.shape[1])]
header
df_LC.columns=header
df_LC.columns
#df_LV_modification
df_LV=d1.iloc[:,np.r_[78:93,271:285]]
header=[i for i in range(0,df_LV.shape[1])]
header
df_LV.columns=header
df_LV.columns
#df_SD_modification
df_SD=d1.iloc[:,285:302]
header=[i for i in range(0,df_SD.shape[1])]
header
df_SD.columns=header
df_SD.columns
cards_user_chose_orderwise_LC=[]
for i in range (2,(df_LC.shape[0])):
  no_of_cards=[] 
  for j in range (0,(df_LC.shape[1])):
    if pd.notna(df_LC.iloc[i,j]):
      temp= int(df_LC.columns[j])
      temp=temp/2
      a=math.floor(temp)
      no_of_cards.append(int(a)+1)
  cards_user_chose_orderwise_LC.append(list(set(no_of_cards)))
cards_user_chose_orderwise_VC=[]
for i in range (2,(df_VC.shape[0])):
  no_of_cards=[] 
  for j in range (0,(df_VC.shape[1])):
    if pd.notna(df_VC.iloc[i,j]):
      temp= int(df_VC.columns[j])
      a=math.floor(temp)
      no_of_cards.append(int(a)+1)
  cards_user_chose_orderwise_VC.append(list(set(no_of_cards)))
k=-1
s=0
S_No=[]
Person_Id=[]
for i in cards_user_chose_orderwise_LC:
  k+=1
  for j in cards_user_chose_orderwise_LC[k]:
    s+=1
    S_No.append(s)
    Person_Id.append(k+1)
d3['S.No.']=S_No
d3['Person_ID']=Person_Id
k=-1
s=0
for i in cards_user_chose_orderwise_LC:
  k+=1
  l=0
  for j in cards_user_chose_orderwise_LC[k]:
    s+=1
    d3.iloc[s-1,71:88] = df_SD.iloc[(k+2),0:]
    d3.iloc[s-1,26:55] = df_LV.iloc[(k+2),0:]
    d3.iloc[s-1,88] = d1.iloc[(k+2),1]
    d3.iloc[s-1,6:23] = d2.iloc[((cards_user_chose_orderwise_LC[k][l])-1),2:]
    d3.iloc[s-1,57:66] = d4.iloc[((cards_user_chose_orderwise_VC[k][l])-1),2:]
    l+=1
s=0
for i in range (2,(df_LC.shape[0])):
  j=0
  while j <= ((df_LC.shape[1])-1):
    if pd.notna(df_LC.iloc[i,j]):
      s=s+1
      if str(df_LC.iloc[i,j])=="Home":
        d3.iloc[(s-1),2] = 1
        if str(df_LC.iloc[i,j+1])=="Leisure":
          d3.iloc[(s-1),3] = 2
        elif str(df_LC.iloc[i,j+1])=="Workplace":
          d3.iloc[(s-1),4] = 2
        elif str(df_LC.iloc[i,j+1])=="Highway":
          d3.iloc[(s-1),5] = 2

      elif str(df_LC.iloc[i,j])=="Leisure":
        d3.iloc[(s-1),3] = 1
        if str(df_LC.iloc[i,j+1])=="Home":
          d3.iloc[(s-1),2] = 2
        elif str(df_LC.iloc[i,j+1])=="Workplace":
          d3.iloc[(s-1),4] = 2
        elif str(df_LC.iloc[i,j+1])=="Highway":
          d3.iloc[(s-1),5] = 2
        
      elif str(df_LC.iloc[i,j])=="Workplace":
        d3.iloc[(s-1),4] = 1
        if str(df_LC.iloc[i,j+1])=="Home":
          d3.iloc[(s-1),2] = 2
        elif str(df_LC.iloc[i,j+1])=="Leisure":
          d3.iloc[(s-1),3] = 2
        elif str(df_LC.iloc[i,j+1])=="Highway":
          d3.iloc[(s-1),5] = 2

      elif str(df_LC.iloc[i,j])=="Highway":
        d3.iloc[(s-1),5] = 1
        if str(df_LC.iloc[i,j+1])=="Home":
          d3.iloc[(s-1),2] = 2
        elif str(df_LC.iloc[i,j+1])=="Leisure":
          d3.iloc[(s-1),3] = 2
        elif str(df_LC.iloc[i,j+1])=="Workplace":
          d3.iloc[(s-1),4] = 2
      j+=2
    else:
      j=j+1
s=0
for i in range (2,(df_VC.shape[0])):
  j=0
  while j <= ((df_VC.shape[1])-1):
    if pd.notna(df_VC.iloc[i,j]):
      s=s+1
      if str(df_VC.iloc[i,j])=="Conventional 4-Wheeler":
        d3.iloc[(s-1),55] = 1
      elif str(df_VC.iloc[i,j])=="Electric 4-Wheeler":
        d3.iloc[(s-1),56] = 1
    j+=1
print(d3)
d3.to_csv("file1.csv")
```
