```
#CODE PREPARED BY GAURAV YASH TIWARI
import pandas as pd
import biogeme.database as db
import biogeme.biogeme as bio
import biogeme.models as models
import biogeme.version as ver
from biogeme.expressions import Beta
url="https://raw.githubusercontent.com/gytrishi/project/main/MNL.csv"
df=pd.read_csv(url)
database=db.Database("testmodel",df)
globals().update(database.variables)

#directly access the names of variables in data file
#from headers import*

#list of parameters
ASC_DA = Beta('ASC_DA',0,None,None,0)
ASC_SR = Beta('ASC_SR',0,None,None,0)
ASC_TR= Beta('ASC_TR',0,None,None,1)
'''BETA_IVTT_DA = Beta('BETA_IVTT_DA',0,None,None,0)
BETA_IVTT_SR = Beta('BETA_IVTT_SR',0,None,None,1)
BETA_IVTT_TR= Beta('BETA_IVTT_TR',0,None,None,1)
BETA_OVTDS_DA = Beta('BETA_OVTDS_DA',0,None,None,0)
BETA_OVTDS_SR = Beta('BETA_OVTDS_SR',0,None,None,1)
BETA_OVTDS_TR= Beta('BETA_OVTDS_TR',0,None,None,1)
BETA_COST_DA = Beta('BETA_COST_DA',0,None,None,0)
BETA_COST_SR = Beta('BETA_COST_SR',0,None,None,1)
BETA_COST_TR= Beta('BETA_COST_TR',0,None,None,1)'''
BETA_IVTT = Beta('BETA_IVTT_DA',0,None,None,0)
BETA_OVTDS = Beta('BETA_OVTDS_DA',0,None,None,0)
BETA_COST = Beta('BETA_COST_DA',0,None,None,0)

#definition of utility function
'''
U1= ASC_DA+\
    BETA_IVTT_DA*IVTT_DA+\
    BETA_OVTDS_DA*OVTDS_DA+\
    BETA_COST_DA*COST_DA
U2= ASC_SR+\
    BETA_IVTT_SR*IVTT_SR+\
    BETA_OVTDS_SR*OVTDS_SR+\
    BETA_COST_SR*COST_SR
U3= ASC_TR+\
    BETA_IVTT_TR*IVTT_TR+\
    BETA_OVTDS_TR*OVTDS_TR+\
    BETA_COST_TR*COST_TR'''
U1= ASC_DA+\
    BETA_IVTT*IVTT_DA+\
    BETA_OVTDS*OVTDS_DA+\
    BETA_COST*COST_DA
U2= ASC_SR+\
    BETA_IVTT*IVTT_SR+\
    BETA_OVTDS*OVTDS_SR+\
    BETA_COST*COST_SR
U3= ASC_TR+\
    BETA_IVTT*IVTT_TR+\
    BETA_OVTDS*OVTDS_TR+\
    BETA_COST*COST_TR


#Associate the Utility with some numbers
U={1:U1,2:U2,3:U3}

#Availability of each alternative as described in the dataset
AV = {1:AV_DA,2:AV_SR,3:AV_TR}


#choice model is a logit model with availability conditions
logprob=models.loglogit(U,AV,CHOICE)

biogeme=bio.BIOGEME(database,logprob)
biogeme.modelName ='testmodel'
results=biogeme.estimate()
pandasresults=results.getEstimatedParameters()
pandasresults
```
